```
import mplfinance as mpf
import matplotlib.pyplot as plt
import yfinance as yf

while True:#增加一層防輸入錯誤確認層
  try:
    stock_num = input("輸入股票代碼:")+".TW"
    print(stock_num)
    dma = yf.Ticker(stock_num).history()
    mpf.plot(dma.tail(30), type='candle', mav=(3,20), volume=True, show_nontrading=True, style='yahoo') #所以下面繪製最後30個交易日的蠟燭圖，並添加成交量與 3、6、9 日均線
  except TypeError:
    print("你輸入了一個錯誤的代碼")
    continue

    
```
